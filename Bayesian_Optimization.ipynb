```python
import numpy as np

def simulated_taVNS_response(amplitude, frequency, pulse_width):
    """
    This simulates a patient's Heart Rate Variability (HRV) improvement.
    Higher score = Better therapeutic effect.
    """
    # Perfect target parameters for this "virtual patient"
    target_amp = 2
    target_freq = 25
    target_width = 200

    # Calculate how far away the AI's guess is from the perfect settings
    distance_from_ideal = (
        (amplitude - target_amp)**2 +
        ((frequency - target_freq) / 10)**2 +
        ((pulse_width - target_width) / 100)**2
    )

    # Base physiological response
    hrv_improvement = 100 - distance_from_ideal

    # Safety penalty: if amplitude is too high, it hurts the patient (drops the score)
    if amplitude > 3.0:
        hrv_improvement -= 50  # Mimics a painful side effect threshold

    return hrv_improvement
```

```python
from bayes_opt import BayesianOptimization

# 1. Define the search boundaries (Safe limits for a typical taVNS device)
search_boundaries = {
    'amplitude': (0.1, 4.0),     # mA (milliamperes)
    'frequency': (1.0, 50.0),    # Hz (hertz)
    'pulse_width': (50.0, 300.0) # microseconds
}

# 2. Initialize the optimizer optimizer
optimizer = BayesianOptimization(
    f=simulated_taVNS_response,
    pbounds=search_boundaries,
    random_state=42, # Keeps results consistent every time you run it
)

# 3. Let it explore!
# init_points = random guesses to start drawing the "blurry map"
# n_iter = how many steps the AI takes to home in on the perfect setting
optimizer.maximize(init_points=3, n_iter=80)

# 4. Print out the ultimate settings found
print("\n--- OPTIMIZATION COMPLETE ---")
print("Best Simulated Target Score Found:", optimizer.max['target'])
print("Optimal Parameters Chosen by AI:", optimizer.max['params'])
```

### Creating an Interactive Slider with `ipywidgets`

Here's how you can make a slider that controls an input to a function in Colab.

```python
import ipywidgets as widgets
from IPython.display import display

# A simple function we want to control with a slider
def calculate_square(value):
    print(f"The square of {value} is {value ** value}")

# Create a slider widget
# min, max, step define the range and granularity
# description is the label for the slider
slider = widgets.IntSlider(
    min=0,
    max=100,
    step=1,
    description='Number:',
    value=50  # Initial value
)

# Link the slider to the function
# Whenever the slider's value changes, calculate_square will be called with the new value
output = widgets.interactive_output(calculate_square, {'value': slider})

# Display the slider and its output
display(slider, output)
```
